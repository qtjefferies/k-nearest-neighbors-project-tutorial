### Import Data 

```python
import pandas as pd 

credits_df = pd.read_csv('tmdb_5000_credits.csv')
movies_df = pd.read_csv('tmdb_5000_movies.csv')

movies_df.head()
```

### SQL Database Storage 

```python
import sqlite3

conn = sqlite3.connect('movies.db')
cursor = conn.cursor()

movies_df.to_sql('movies', con = conn, if_exists='replace', index=False)
credits_df.to_sql('credits', con = conn, if_exists='replace', index=False)

conn.commit()

```

```python
movies_db = pd.read_sql_query("Select * FROM movies", conn)
movies_db.columns
```

```python
credits_db = pd.read_sql_query("Select * FROM credits", conn)
credits_db.head()
```

##### Join tables together SQL


```python
query = '''
    SELECT c.movie_id, m.title, m.overview, m.genres, m.keywords, c.cast, c.crew
    FROM movies m
    INNER JOIN credits c
    ON c.title = m.title
    
'''

joined_db = pd.read_sql_query(query, conn)
joined_db.head()
```

```python
# Data transform as expected
import json

def load_json_safe(json_str, default_value = None):
    try:
        return json.loads(json_str)
    except (TypeError, json.JSONDecodeError):
        return default_value
    
```

### convert json fields to name

```python
joined_db["genres"] = joined_db["genres"].apply(lambda x: [item["name"] for item in json.loads(x)] if pd.notna(x) else None)
joined_db["keywords"] = joined_db["keywords"].apply(lambda x: [item["name"] for item in json.loads(x)] if pd.notna(x) else None)

joined_db["cast"] = joined_db["cast"].apply(lambda x: [item["name"] for item in json.loads(x)][:3] if pd.notna(x) else None)

joined_db["crew"] = joined_db["crew"].apply(lambda x: [crew_member["name"] for crew_member in json.loads(x)] if pd.notna(x) else None)

joined_db["overview"] = joined_db["overview"].apply(lambda x: [x])

joined_db.head()
```

```python
def removeSpaces(tag):
    return tag.apply(lambda x: [[item.replace(' ', '') for item in x] if isinstance(x, list) else x])

```

```python
def removeSpaces(tag):
    return tag.apply(lambda x: [[str(item).replace(' ', '') for item in x] if isinstance(x, list) else x] if x is not None else x)

joined_db['genres'] = removeSpaces(joined_db['genres'])
joined_db['cast'] = removeSpaces(joined_db['cast'])
joined_db['crew'] = removeSpaces(joined_db['crew'])
joined_db['keywords'] = removeSpaces(joined_db['keywords'])
joined_db['overview'] = removeSpaces(joined_db['overview'])

print(joined_db.head(5))
```

### Add Tag Field

```python

def combineList(df):
    combined = joined_db['crew'] + joined_db['genres'] + joined_db['keywords'] + joined_db['cast'] 
    return combined

print(combineList(joined_db))

joined_db['tags'] = combineList(joined_db)

# Set display options to see all values in one row
#pd.set_option('display.max_colwidth', None)

print(joined_db['tags'][10])


```

### Build KNN Model

#### Vectorize Tags

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Convert list of tags to a single string
joined_db['tags'] = joined_db['tags'].apply(lambda x: ' '.join(map(str, x)) if isinstance(x, list) else '')

Tfidf = TfidfVectorizer(stop_words='english')
tags_matrix = Tfidf.fit_transform(joined_db['tags'])

print(tags_matrix)
```

#### Cosine Similarity 

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(tags_matrix)
# print(cosine_sim)
```

```python
from sklearn.neighbors import NearestNeighbors

knn = NearestNeighbors(n_neighbors = 6, algorithm ="brute", metric="cosine")
knn.fit(tags_matrix)

def get_movie_recommendations(movie_title):
    movie_index = joined_db[joined_db["title"] == movie_title].index[0]
    distances, indices = knn.kneighbors(tags_matrix[movie_index])
    similar_movies = [(joined_db["title"][i], distances[0][j]) for j, i in enumerate(indices[0])]
    return similar_movies[1:]

input_movie = "Avatar"
recommendations = get_movie_recommendations(input_movie)
print("Film recommendations '{}'".format(input_movie))
for movie, distance in recommendations:
    print("- Film: {}".format(movie))
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Reduce the dimensionality of the TF-IDF matrix to 2 components using PCA
pca = PCA(n_components=2)
reduced_matrix = pca.fit_transform(tags_matrix.toarray())  # Convert sparse matrix to dense

# Plot the results
plt.figure(figsize=(10, 8))
plt.scatter(reduced_matrix[:, 0], reduced_matrix[:, 1], alpha=0.7)

# Optionally, label the points by their index
for i in range(100):
    plt.text(reduced_matrix[i, 0], reduced_matrix[i, 1], str(i), fontsize=12)

plt.title('PCA Plot of Movies Based on Cosine Similarity')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```
